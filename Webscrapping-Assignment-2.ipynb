# Q.7-Scrapping of 100 "Shoes" data by  "Price filter" and Color filter to “Black” from "www.myntra.com/shoes" along with "Brand”,"Short Shoe description" and "Price of the shoe"

```python
from selenium import webdriver
import pandas as pd
import time
import warnings
warnings.filterwarnings('ignore')

driver=webdriver.Chrome(r'C:\Users\clickmoney\Downloads\Compressed\Flip Robo\chromedriver_win32\chromedriver.exe')

driver.get("https://www.myntra.com/shoes")

#applying price filter
price_filter=driver.find_element(By.XPATH,"//ul[@class='price-list']/li[2]/label/div")
price_filter.click()
time.sleep(5)

#applying filter color black
color_filter=driver.find_element(By.XPATH,'//li[@class="colour-listItem"]/label/div')
color_filter.click()
time.sleep(5)

shoe_brand=[]
shoe_description=[]
shoe_price=[]
def scrap_shoes(driver):
    brand=driver.find_elements(By.XPATH,"//div[@class='product-productMetaInfo']/h3")
    for i in brand:
        shoe_brand.append(i.text)

    description=driver.find_elements(By.XPATH,"//div[@class='product-productMetaInfo']/h4[1]")
    for i in description:
        shoe_description.append(i.text)

    price=driver.find_elements(By.XPATH,"//div[@class='product-productMetaInfo']/div/span[1]")
    for i in price:
        shoe_price.append(i.text)

scrap_shoes(driver)        

while((len(shoe_brand) < 100) and (len(shoe_description)<100) and (len(shoe_price)<100)):
     next_page=driver.find_element(By.XPATH,'//a[@rel="next"]')
     next_page.click()
     time.sleep(5)    
     page=driver.find_element(By.XPATH,'//li[@class="pagination-active"]/a')
     url=page.get_attribute('href')
     driver.get(url)
     time.sleep(5)
     scrap_shoes(driver)

#print(len(shoe_brand),len(shoe_description),len(shoe_price))

df=pd.DataFrame({"Shoe Brand":shoe_brand,"Short Shoe Description":shoe_description,"Price":shoe_price})
df
```

# Q.10-Python program to scrape the salary data for "Data Scientist" designation with "Company_name", "Number_of_salaries", "Average_salary", "Minimum_Salary", "Maximum_salary"

```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time

#Lets first connect ro webdriver
driver=webdriver.Chrome(r'C:\Users\clickmoney\Downloads\Compressed\Flip Robo\chromedriver_win32\chromedriver.exe')

#Opening target url page on automated chrome browser
driver.get("https://www.ambitionbox.com/")

#Click salaries
salaries=driver.find_element(By.XPATH,"/html/body/div/div/div/div[1]/header/nav/ul/li[3]/span")
salaries.click()

#Click salaries
browse=driver.find_element(By.XPATH,"/html/body/div/div/div/div[1]/header/nav/ul/li[3]/div/ul/li[1]/div/div[2]/p")
browse.click()

#enter Data Scientist in input search bar
search=driver.find_element(By.XPATH,"/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/input")
search.send_keys("Data Scientist")
time.sleep(5)

#manually selected designation 

company_name=[]
name=driver.find_elements(By.XPATH,'//div[@class="company-info-wrapper"]')
for i in name:
    company_name.append(i.text)
    
time.sleep(2)

# code for scraping minimum salaries using xpath
minimum_salary=[]
minimum_sal=driver.find_elements(By.XPATH,'//div[@class="salary-values"]/div[1]')        
for i in minimum_sal:
    minimum_salary.append(i.text)

time.sleep(2)

average_salary=[]
average_sal=driver.find_elements(By.XPATH,"//div[@class='solid-salary-bar salary-bar-wrapper']/div/p")  
for i in average_sal:
    average_salary.append(i.text)

time.sleep(2)

# code for scraping maximum salaries using xpath
maximum_salary=[]
maximum_sal=driver.find_elements(By.XPATH,"//div[@class='salary-values']/div[2]")        
for i in maximum_sal:
    maximum_salary.append(i.text)

time.sleep(2)

# code for scraping total salaries using xpath
total_salaries=[]
ttl_sal=driver.find_elements(By.XPATH,"//div[@class='company-info']/div/span")        
for i in ttl_sal:
    total_salaries.append(i.text.replace('.',''))    

time.sleep(2)

#checking the eqality of length
print(len(company_name),len(total_salaries),len(average_salary),len(minimum_salary),len(maximum_salary))
#converting into Dataframe (only 10 Records)
df=pd.DataFrame({"Company Name":company_name[:10],"Total Salaries":total_salaries[:10],"Average_Salary":average_salary[:10],"Minimum Salary":minimum_salary[:10],"Maximum Salary":maximum_salary[:10]})
df
```
